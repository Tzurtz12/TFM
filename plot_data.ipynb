## Cell to plot 2D data: 
* Barrier plots
* Gate effect
* Coulomb Oscillations

```python
# Step 1: Read the data
# Replace 'data.csv' with your actual file name
df = pd.read_csv('./GRAPHS/mosfet_28k.txt', header=None, names=['x', 'y'],sep=' ')

# Step 2: Plot the data
plt.style.use('fast')  # Optional: Use a predefined style for aesthetics
plt.figure(figsize=(10, 6))  # Set figure size for better visibility

plt.plot(df['x'], df['y'], marker='o', linestyle='-', color='royalblue', markersize=8, linewidth=2)

plt.title('DAC vs ADC', fontsize=16)  # Add a title
plt.xlabel('DAC 3', fontsize=14)  # Label x-axis
plt.ylabel('ADC 3', fontsize=14)  # Label y-axis
plt.grid(True)  # Show grid for easier reading

# Optional: Customize ticks for a more detailed scale
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.show()  # Display the plot
```

## Cell for Coulomb diamonds

```python

# Step 1: Read the data
df = pd.read_csv('./GRAPHS/diamonds1.txt', header=None, names=['x', 'y', 'values'], sep=' ')

# Step 2: Interpolate the values onto a grid
# Create grid coordinates
xi = np.linspace(df['x'].min(), df['x'].max(), 100)
yi = np.linspace(df['y'].min(), df['y'].max(), 100)
xi, yi = np.meshgrid(xi, yi)

# Interpolate
zi = griddata((df['x'], df['y']), df['values'], (xi, yi), method='cubic')

# Step 3: Plot the data
plt.style.use('fast')
plt.figure(figsize=(10, 6))

# Use contourf to plot the interpolated values
plt.contourf(xi, yi, zi, levels=100, cmap='viridis')

plt.colorbar(label='Values')  # Adds a color bar to indicate the scale of values

plt.title('DAC vs ADC with Gradient', fontsize=16)
plt.xlabel('DAC 3', fontsize=14)
plt.ylabel('ADC 3', fontsize=14)
plt.grid(False)

plt.show()
```
