## Cell for Coulomb diamonds

```python


df = pd.read_csv('./GRAPHS/SET/experiment_273.csv', header=None, names=['x', 'y', 'values'], sep='\t')


xi = np.linspace(df['x'].min(), df['x'].max(), 100)
yi = np.linspace(df['y'].min(), df['y'].max(), 100)
xi, yi = np.meshgrid(xi, yi)


zi = griddata((df['x'], df['y']), df['values']*1e12, (xi, yi), method='cubic')

max = 3.5
zi = np.clip(zi, 0, max)  

plt.style.use('fast')
plt.figure(figsize=(10, 6))


plt.contourf(yi, xi, zi, levels=100, cmap='viridis')


cbar = plt.colorbar(label='I$_{sd}$(pA)')
cbar.ax.set_ylabel('I$_{sd}$(pA)', fontsize=18) 
#plt.title('DAC vs ADC with Gradient', fontsize=16)
plt.xlabel('V$_g$ (V)', fontsize=20)
plt.ylabel('V$_{sd}$(V)', fontsize=20)
plt.grid(False)

plt.show()
```
