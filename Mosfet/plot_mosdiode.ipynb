```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import griddata
```

```python


df = pd.read_csv('./old/3_PLC.txt', header=None, names=['x', 'y1','y2','y3'], sep='\t')

plt.style.use('fast')  
fig, ax1 = plt.subplots(figsize=(9, 7))


color = 'tab:red'
ax1.set_xlabel('V$_{ds}$ (V)')
ax1.set_ylabel('Source and Drain current (nA)', color=color)
ax1.plot(df['x'], df['y1']*1e9, color='royalblue', linewidth=2, label='Drain current')
ax1.plot(df['x'], df['y2']*1e9, color='red', linewidth=2, label='Source current')
ax1.tick_params(axis='y', labelcolor=color)

# Create a second y-axis for the third dataset
ax2 = ax1.twinx()  
color = 'tab:green'
ax2.set_ylabel('Gate current (pA)', color=color)  # we already handled the x-label with ax1
ax2.plot(df['x'], df['y3']*1e12, color='green', linewidth=2, label='Gate current')
ax2.tick_params(axis='y', labelcolor=color)
ax1.grid(True, which='both', linestyle='--', linewidth=0.5)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
ax1.legend(loc='upper left', fontsize=12)
ax2.legend(loc='upper right', fontsize=12)

plt.legend(fontsize=12)
plt.show()  
```
